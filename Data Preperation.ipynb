```python
data_dict_2022 = df_2022.loc[0,:].to_dict()
# data_dict_2022
```

```python
def extract_mcq_questions(data_dict, q_string, descriptor_string):
    q_keys = list(filter(lambda x: q_string in x, list(data_dict.keys())))
    mcq_dict = {}
    for q_key in q_keys:
        mcq_dict[q_key] = descriptor_string + re.search('Selected Choice - (.+?)$', data_dict[q_key]).group(1).strip()
    return mcq_dict
```

```python
q_descriptor_mapping_2022 = [( "Q6_", "MOOCPlatforms_"),( "Q7_", "IntroToDSPlatform_"),( "Q12_", "ProgLanguage_"),( "Q13_", "IDE_"),( "Q14_", "HostedNB_"),( "Q15_", "VizLib_"),( "Q17_", "MLFramework_"),( "Q18_", "MLAlgo_"),( "Q19_", "CVAlgo_"),( "Q20_", "NLPAlgo_"),( "Q21_", "DownloadPreTrainedWeights_"),( "Q31_", "CloudProviderUsed_"),( "Q33_", "CloudComputeUsed_"),( "Q34_", "CloudStorageUsed_"),( "Q35_", "DBUsed_"),( "Q36_", "BIToolUsed_"),( "Q37_", "MLaaS_"),( "Q38_", "AutoMLUsed_"),( "Q39_", "MLDeploymentToolsUsed_"),( "Q40_", "MLMonitoringToolUsed_"),( "Q41_", "ResponsibleAIToolUsed_"),( "Q42_", "HardwareUsed_"),( "Q44_", "DSContentMediaConsumed_")] 
```

```python
map_cols_2022 = {
    "Q2": "Age",
    "Q3": "Gender",
    "Q4": "Country",
    "Q5": "IsStudent",
    "Q8": "HighestEducation",
    "Q9": "HasPublishedResearch",
    "Q10_1": "TheoreticalResearch",
    "Q10_2": "AppliedResearch",
    "Q10_3": "NonMLResearch",
    "Q11": "Programming Experience",
    "Q16": "MLExposure",
    "Q22": "MLReposUsed",
    "Q23": "CurrentTitle",
    "Q24": "Industry",
    "Q25": "CompanySize",
    "Q26": "DSTeamSize",
    "Q27": "IsMLUsedAtWork",
    "Q29": "AnnualComp(USD)",
    "Q30": "InvestmentInML(USD)",
    "Q32": "BestCloudProvider",
    "Q43": "TPUUsageFrequency"
}
mcq_columns = []
for q,d in q_descriptor_mapping_2022:
    q_d_mapping = extract_mcq_questions(data_dict_2022, q, d)
    map_cols_2022 = map_cols_2022 | q_d_mapping
    mcq_columns.extend(list(q_d_mapping.values()))
```

```python
df_2022 = df_2022.rename(map_cols_2022, axis=1)
df_2022 = df_2022.loc[1:,:]
```

```python
for mcq_column in mcq_columns:
    df_2022.loc[~df_2022[mcq_column].isin([np.nan]), mcq_column] = 1
    df_2022.loc[df_2022[mcq_column].isin([np.nan]), mcq_column] = 0
```

```python
def categorize_education(education:str) -> str:
    """Assigns more general categories to education levels.
    """
    if education in [
        "No formal education past high school", 
        "Some college/university study without earning a bachelors degree"
    ]:
        return "Lower than Bachelor"
    
    if education == "Bachelors degree":
        return "Bachelor"
    
    if education == "Masters degree":
        return "Master"
    
    if education in ["Doctoral degree", "Professional doctorate"]:
        return "Higher than Master"
    
    return "Other"


def assign_label(service:str):
    """It returns the company name to which the product belongs. 
    It takes care only of the 3 big techs: Google, Microsoft, Amazon.
    """
    
    if "google" in service.lower():
        return "Google"
    
    if "aws" in service.lower() or "amazon" in service.lower():
        return "Amazon"
    
    if "azure" in service.lower() or "microsoft" in service.lower():
        return "Microsoft"
    
    if "ibm" in service.lower():
        return "IBM"

    return "Other"
```

```python
data["HighestEducation"] = data["HighestEducation"].apply(lambda x: categorize_education(x))
```
